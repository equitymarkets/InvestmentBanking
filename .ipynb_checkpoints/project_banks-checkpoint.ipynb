```python
x_axis = np.sort(bank_data_df['Fiscal Year'].unique(), axis=-1)

x_tick_locations = [value for value in x_axis]
plt.figure(0,figsize=(10,6))

#marker ='o',  for points
jeff = plt.plot(x_axis, jeff_plot_df['Total IB Fees'], color='blue', label="Jefferies")
#plt.text('Jefferies')
    #plt.text(0.9, 0.2, 'cos')
suisse = plt.plot(x_axis, suisse_plot_df['Total IB Fees'], color='green', label="Credit Suisse")
citi = plt.plot(x_axis, citi_plot_df['Total IB Fees'], color='red', label="CitiBank")
bar = plt.plot(x_axis, bar_plot_df['Total IB Fees'], color='purple', label="Barclays")
gs = plt.plot(x_axis, gs_plot_df['Total IB Fees'], color='darkblue', label="Goldman Sachs")
jp = plt.plot(x_axis, jp_plot_df['Total IB Fees'], color='pink', label="JP Morgan")
rbc = plt.plot(x_axis, rbc_plot_df['Total IB Fees'], color='yellow', label="Royal Bank of Canada")
d = plt.plot(x_axis, deutsche_plot_df['Total IB Fees'], color='orange', label="Deutsche Bank")
bac = plt.plot(x_axis, bac_plot_df['Total IB Fees'], color='darkred', label="Bank of America")
ms = plt.plot(x_axis, ms_plot_df['Total IB Fees'], color='darkgreen', label="Morgan Stanley")
plt.xticks(x_tick_locations)
legend = plt.legend(loc='lower right')
plt.title("Investment Banking Fees")
plt.xlabel("Fiscal Year")
plt.ylabel("Investment Banking Fees (millions)")
plt.grid()
plt.show()
```

```python
bank_list = ['Jefferies', 'Credit Suisse', 'Citi Bank', 'Barclays',
       'Goldman Sachs', 'JP Morgan', 'RBC Bank', 'Deutsche Bank',
       'Bank of America', 'Morgan Stanley']
bank_pie = np.array([jeff_plot_df['Total IB Fees'].sum(), suisse_plot_df['Total IB Fees'].sum(), citi_plot_df['Total IB Fees'].sum(), bar_plot_df['Total IB Fees'].sum(),
                     gs_plot_df['Total IB Fees'].sum(), jp_plot_df['Total IB Fees'].sum(), rbc_plot_df['Total IB Fees'].sum(), deutsche_plot_df['Total IB Fees'].sum(), bac_plot_df['Total IB Fees'].sum(),
                     ms_plot_df['Total IB Fees'].sum() ])
plt.pie(bank_pie, labels=bank_list, autopct="%1.1f%%", shadow=True, startangle=60)
plt.figure(1)
plt.title("Investment Banking Fees, 2017-2021 Cumulative")
plt.show()
```

```python
revenue_2021 = bank_data_df.groupby('Bank').head(1)
bank_list = []
revenue_list = []
for row in revenue_2021.iterrows():
    bank_list = revenue_2021['Bank']
    revenue_list = revenue_2021['Total Revenue']
TR_pie = np.array(revenue_list)
plt.pie(TR_pie, labels=bank_list, autopct="%1.1f%%", shadow=True, startangle=60)
plt.title("Market Share of Total Revenue, 2021")
plt.show()
```

```python
assets_2021 = bank_data_df.groupby('Bank').head(1)
bank_list = []
assets_list = []
for row in assets_2021.iterrows():
    bank_list = assets_2021['Bank']
    assets_list = assets_2021['Total Assets']
assets_pie = np.array(assets_list)
plt.pie(assets_list, labels=bank_list, autopct="%1.1f%%", shadow=True, startangle=60)
plt.title("Market Share of Total Assets, 2021")
plt.show()
```
